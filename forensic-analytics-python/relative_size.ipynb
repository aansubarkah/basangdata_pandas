```python
sales = pd.read_csv('https://github.com/aansubarkah/basangdata_pandas/raw/master/AW/Sales.SalesOrderHeader.csv', sep='\t')
```

# Gunakan data dengan nilai $\ge$ 1

```python
sales[sales['SubTotal'] < 1]
```

Karena tidak terdapat data $\lt$ 1 maka semua data akan digunakan

# Kelompokkan data dan hitung berapa kemunculan

```python
salesgr1 = sales[['CustomerID', 'SubTotal']].groupby('CustomerID', as_index=False).count()
```

```python
salesgr1.sort_values(by='SubTotal', ascending=False)
```

# Gunakan hanya CustomerID yang jumlah kemunculannya $\gt$ 1

```python
salesgr2 = salesgr1[salesgr1['SubTotal'] > 1]
salesgr2.shape
```

# Ambil Max 1-2 SubTotal

```python
sales2 = sales[sales['CustomerID'].isin(salesgr2['CustomerID'])][['CustomerID', 'SubTotal']]
```

```python
sales2.head(5)
```

```python
grouped = sales2.groupby(['CustomerID'])['SubTotal']
```

```python
sales2['first_highest'] = grouped.transform(lambda x: x.nlargest(1).max())
sales2['second_highest'] = grouped.transform(lambda x: x.nlargest(2).values[-1])
sales2.drop_duplicates(subset='CustomerID', inplace=True)
sales2['relative_size'] = sales2['first_highest'] / sales2['second_highest']
```

```python
sales2.sort_values(by='relative_size', ascending=False)[['CustomerID', 'first_highest', 'second_highest', 'relative_size']]
```

```python
sales2.sort_values(by='relative_size', ascending=False)[['CustomerID', 'first_highest', 'second_highest', 'relative_size']]
```
