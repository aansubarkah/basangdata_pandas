# Impor Library Diperlukan

```python
import pandas as pd
import re
```

# Load Data

```python
df = pd.read_csv('https://raw.githubusercontent.com/aansubarkah/basangdata_pandas/master/pandas-beginner/penjualan.csv')
```

```python
df
```

# Memilih Satu Kondisi

## Memilih Angka

### Memilih Sama Dengan

```python
df[df['jan'] == 200]
```

```python
df[df.jan.eq(200)]
```

### Memilih Tidak Sama Dengan

```python
df[df['jan'] != 200]
```

```python
df[df.jan.ne(200)]
```

```python
df[~(df['jan'] == 200)]
```

### Memilih Lebih Dari

```python
df[df['jan'] > 200]
```

```python
df[df.jan.gt(200)]
```

### Memilih Lebih Dari atau Sama Dengan

```python
df[df['jan'] >= 200]
```

```python
df[df.jan.ge(200)]
```

### Memilih Kurang Dari

```python
df[df['jan'] < 200]
```

```python
df[df.jan.lt(200)]
```

### Memilih Kurang Dari atau Sama Dengan

```python
df[df['jan'] <= 200]
```

```python
df[df.jan.le(200)]
```

## Memilih Teks

### Memilih Mengandung Teks

```python
df[df['nama'].str.contains('Burhan')]
```

```python
df[df['nama'].str.contains('Jul')]
```

### Abaikan Huruf Besar/Kecil

```python
df[df['nama'].str.contains('in')]
```

```python
import re
df[df['nama'].str.contains('in', flags=re.IGNORECASE)]
```

## Memilih Data yang Berada di List/Dataframe Lain

```python
jan_sales = [95, 125]
df[df['jan'].isin(jan_sales)]
```

```python
d = {'tinggi': [320, 300], 'rendah': [95,100]}
df2 = pd.DataFrame(data=d)
df2
```

```python
df[df['jan'].isin(df2['tinggi'])]
```

```python
df[df['jan'].isin(df2['rendah'])]
```

# Memilih Lebih dari Satu Kondisi

## Memilih Diantara Dua Rentang Nilai

```python
df[(df['jan'] > 100) & (df['jan'] < 300)]
```

## or Condition

```python
df[(df['nama'].str.contains('on', flags=re.IGNORECASE)) | (df['jan'] > 200)]
```
