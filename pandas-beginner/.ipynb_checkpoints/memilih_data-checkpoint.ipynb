# Menampilkan Hanya Kolom Tertentu

**.loc** Memilih kolom dan baris berdasarkan label.<br>
**.iloc** Memilih baris dan kolom berdasarkan posisi, yang ditunjukkan dengan angka integer<br>
**[]** Memilih kolom dan baris berdasarkan nama kolom

***

Menampilkan baris dengan indeks berlabel 0

```python
df.loc[0]
```

```python
df.iloc[0]
```

```python
df[:1]
```

***

Menampilkan **baris** indeks berlabel **0** dan **kolom** berlabel **jan**

```python
df.loc[0, 'jan']
```

```python
df.iloc[0, 1]
```

```python
df[['jan']][:1]
```

```python
df[:1][['jan']]
```

***

Menampilkan baris berlabel 0 sampai 3 dan kolom **nama**, **jan**, **mar**

```python
df.loc[0:3, ['nama', 'jan', 'mar']]
```

```python
df.iloc[0:3, [0, 1, 3]]
```

```python
df[['nama', 'jan', 'mar']][:3]
```

```python
df[:3][['nama', 'jan', 'mar']]
```

***

Menampilkan semua baris pada kolom **nama** dan **apr**

```python
df.loc[:, ['nama', 'apr']]
```

```python
df.iloc[:, [0, 4]]
```

```python
df[['nama', 'apr']]
```

***

Menampilkan semua baris pada kolom nama sampai feb

```python
df.loc[:, 'nama':'feb']
```

```python
df.iloc[:, :3]
```

```python
df[['nama', 'jan', 'feb']]
```

***

Menampilkan semua baris dan semua kolom

```python
df.loc[:, :]
```

```python
df.loc[:,]
```

```python
df.iloc[:, :]
```

```python
df.iloc[:]
```

```python
df
```

Tidak mendefinisikan baris yang akan diseleksi tidak dapat dilakukan

```python
df.loc[,:]
```

```python
df.iloc[,:]
```
